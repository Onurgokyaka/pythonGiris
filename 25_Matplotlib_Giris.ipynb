```python
numpyDizisi1=np.linspace(0,10,20) # 0-10 arasi 20 elemanl dizi,eit aralkl
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1 **3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b--") 
plt.xlabel("X")
plt.ylabel("X zeri 3")
plt.title("Saylarn kpleri ile arasndaki iliki")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"r*-") 
plt.xlabel("X")
plt.ylabel("X zeri 3")
plt.title("Saylarn kpleri ile arasndaki iliki")
```

```python
plt.subplot(1,2,1) # 1 sra 2 kolon olacak, 1. kolon bu
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1 sra 2 kolon olacak, 2. kolon bu
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure=plt.figure()
figureAxes= benimFigure.add_axes([0.2,0.2,0.9,0.9]) # son ikili x ekseni v ye ekseni lleri  
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g*-")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik basl")
```

```python
figure2=plt.figure()

axes1= figure2.add_axes([0.1,0.1,0.9,0.9]) # son ikili x ekseni v ye ekseni lleri  
axes2= figure2.add_axes([0.2,0.6,0.3,0.3]) # ieri giren grafik, ilk iki deerle oynayarak yerini dzenleriz

axes1.plot(numpyDizisi1,numpyDizisi2,"r*-")
axes1.set_xlabel("X ekseni veri ismi")
axes1.set_ylabel("Y ekseni veri ismi")
axes1.set_title("Grafik basl")

axes2.plot(numpyDizisi2,numpyDizisi1,"b--")
axes2.set_xlabel("X ekseni kk")
axes2.set_ylabel("Y ekseni kk")
axes2.set_title("Grafik bal")
```

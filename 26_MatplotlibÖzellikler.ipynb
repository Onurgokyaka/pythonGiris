```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1=np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1 ** 2
```

```python
numpyDizisi2
```

```python
(benimFigure,benimEksen)=plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2, "#B75B00",alpha=0.3) # alpha değeri 0 a yalastıkca saydamlasıyor, yani transparan
benimEksen.plot(numpyDizisi2,numpyDizisi1,"#975EC3") # renk kodu girebiliriz bu şekilde,rgb to hex
```

```python
(yeniFigur,yeniEksen)=plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi1+2,color="blue",linewidth=5.0)  # linewith ile çizgi kalınlıgını değiştirdik
yeniEksen.plot(numpyDizisi1,numpyDizisi1+3,color="yellow",linestyle=":")

yeniEksen.plot(numpyDizisi1,numpyDizisi1+4,"#2DE045",linestyle="--")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+5,"black",linestyle="-.",marker="*",markersize=8,markerfacecolor="red") 
yeniEksen.plot(numpyDizisi1,numpyDizisi1+6,"#2DE0A0",linestyle="--",marker="o",markersize=4,markerfacecolor="yellow")
```

```python
yeniFigur.savefig("benimfigur2.png", dpi=200)  # grafikten sonra bunu girip pnd olarak net sekilde alırız
```

```python
y=dataFrame["Type"].values  # bunu bir numpy dizisine cevirdik
x=dataFrame.drop("Type", axis = 1 ).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping  # ezberlemeyi engellemek için
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu")) # sütun,kriter sayısı kadar units 
model.add(Dense(units=15,activation = "relu"))  # ara nöronları 30 dan düşük deneyelim ilk
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid")) # sonuncu 1 degerlı olmalı

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=700, validation_data=(x_test,y_test),verbose=1)
```

```python
model.history.history
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot() # grafikte ezberleme olduğu aşikar
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu")) # sütun,kriter sayısı kadar units 
model.add(Dense(units=15,activation = "relu"))  # ara nöronları 30 dan düşük deneyelim ilk
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid")) # sonuncu 1 degerlı olmalı

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train,
          y=y_train,
          epochs=700,
          validation_data=(x_test,y_test),
          verbose=1,
          callbacks=[earlyStopping])
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot() # ezberleme olan grafiğe göre ciddi iyileşme var
```

## Dropout (layerlarda ezberleme yani overfitting varsa onları atmaya yarar,over fitting varsa kullanmamız iyi olur)

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu")) # sütun,kriter sayısı kadar units 
model.add(Dropout(0.5))  # 0.5 falan iyidir, 1 ve 0 arasıdır

model.add(Dense(units=15,activation = "relu"))  # ara nöronları 30 dan düşük deneyelim ilk
model.add(Dropout(0.5))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.5))

model.add(Dense(units=1,activation = "sigmoid")) # sonuncu 1 degerlı olmalı

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,
          y=y_train,
          epochs=700,
          validation_data=(x_test,y_test),
          verbose=1,
          callbacks=[earlyStopping])
```

```python
lossdf=pd.DataFrame(model.history.history)
```

```python
lossdf.plot()
```

```python
tahminlerimiz=model.predict_classes(x_test)  # verilen girdilere göre yeni tahminleri çıkar
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,tahminlerimiz))
# 0 ları yüzde 90 1 leri yüzde 93 oranla tahmin etmiş
```

```python
print(confusion_matrix(y_test,tahminlerimiz))
```
